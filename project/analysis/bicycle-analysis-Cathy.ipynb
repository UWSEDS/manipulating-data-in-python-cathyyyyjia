## 1
**Read the CSV file into a pandas dataframe. (1 pt)**

```python
import pandas as pd
bicycle = pd.read_csv('../data/bicycle.csv')
bicycle.head(10)
```

## 2
**Add columns to the dataframe containing: (3 pt)**
#### i. The total (East + West) bicycle count

```python
# Total bicycle count = eEast bicycle count + West bicycle count
bicycle['Total Bicycle Count'] = bicycle['Fremont Bridge East Sidewalk'] + bicycle['Fremont Bridge West Sidewalk']
bicycle.head(10)
```

#### ii. The hour of the day

```python
# Convert 'Date' from string to datetime
dates = pd.to_datetime(bicycle['Date'])
# Extract hour from 'Date'
bicycle['Hour'] = dates.dt.hour
bicycle.head(10)
```

#### iii. The year

```python
# Extract year from 'Date'
bicycle['Year'] = dates.dt.year
bicycle.head(10)
```

## 3
**Create a dataframe with the subset of data from the year 2016 (1 pt)**

```python
# Subset the bicycle dataframe where the year is 2016
bicycle_2016 = bicycle[bicycle['Year']==2016]
bicycle_2016.head(10)
```

## 4
**Use pandas + matplotlib to plot the counts by hour. (i.e. hour of the day on the x-axis, total daily counts on the y-axis) (1 pt)**

```python
# Group bicycle dateframe by 'Hour'
# Compute the sum of total bicycle counts
byHour = bicycle.groupby(['Hour']).sum()
byHour.head(10)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# Plot the total daily counts by hour of the day
temp = plt.bar(byHour.index, byHour['Total Bicycle Count'])
temp = plt.xlabel('Hour')
temp = plt.ylabel('Counts')
temp = plt.title('Total Daily Counts by Hour of the Day')
plt.show()
```

## 5
**Use pandas to determine what is (on average) the busiest hour of the day (1 pt)**

```python
# Group bicycle dateframe by 'Hour'
# Compute average total bicycle counts
# Sort by average total bicycle counts in descending order
# Display the top one 'Hour'
bicycle.groupby(['Hour']).mean().sort_values('Total Bicycle Count', ascending=False).head(1)
```

Therefore, the busiest hour of the day on average is 5PM.
